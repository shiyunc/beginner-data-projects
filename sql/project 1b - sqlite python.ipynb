## Sqlite + Python: upload csv using jupyter notebook

```python
# import libraries
import pandas as pd

# to create database
import sqlite3 as db
```

```python
# load data file
df = pd.read_csv("Iris.csv")

# read data
df
```

### Sqlite connect to database

> No database

```python
# create and connect to a database

connection = db.connect('uploadcsv.db')
```

```python
# df.to_sql('table_name', connection_name) to create a table in database to run queries
# load data to sqlite: fail, replace, append

df.to_sql('table2', connection, if_exists='replace')
```

```python
# load sql to ipython

%load_ext sql
```

```python
# refer to the database

%sql sqlite:///uploadcsv.db
```

```python
%%sql

SELECT * FROM table2;
```
