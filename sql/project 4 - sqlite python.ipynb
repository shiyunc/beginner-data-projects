## Sqlite + Python: sql analysis using jupyter notebook (to be continued)

```python
# import libraries
import pandas as pd

# to create database
import sqlite3 as db
```

```python
# load data file
df = pd.read_csv("world_population.csv")

# read data
df
```

```python
# create and connect to a database

connection = db.connect('population.db')
```

```python
# df.to_sql('table_name', connection_name) to create a table in database to run queries
# load data to sqlite: fail, replace, append

df.to_sql('table_population', connection, if_exists='replace')
```

```python
# load sql to ipython

%load_ext sql
```

```python
# refer to the database

%sql sqlite:///population.db
```

```python
%%sql

SELECT *
  FROM sqlite_master
  WHERE type='table'; 
```

```python
%%sql

SELECT * FROM table_population;
```

### Q: Find the maximum and minimum of population 2022 and growth rate in all the countries?

```python
%%sql
SELECT MIN(Population_2022) AS min_population,
       MAX(Population_2022) AS max_population,
       MIN(GrowthRate) AS min_growthrate,
       MAX(GrowthRate) max_growthrate 
  FROM table_population;
```

### Q: Find the minimum growth rate?

```python
%%sql
SELECT *
  FROM table_population
  WHERE GrowthRate == (SELECT MIN(GrowthRate) FROM table_population);
```

### Q: Find the maximum growth rate?

```python
%%sql
SELECT *
  FROM table_population
  WHERE GrowthRate == (SELECT MAX(GrowthRate) FROM table_population);
```

### Q: Find the average growth rate and area?

```python
%%sql

SELECT AVG(GrowthRate) AS avg_GrowthRate, AVG(area) AS avg_area
  FROM table_population
```

### Q: Find the (ascending from lowest to highest) rank of the countries based on the average population 2022 and area?

```python
%%sql
SELECT Rank, CountryTerritory, Population_2022, GrowthRate, Area, WorldPopulationPercentage
  FROM table_population
  WHERE Population_2022 > (SELECT AVG(Population_2022) 
                           FROM table_population)

  AND area < (SELECT AVG(area) 
              FROM table_population)

  ORDER BY 1 ASC;
```
